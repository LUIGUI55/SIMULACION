# Regresion lineal: costo de un incidente de seguridad

En este caso se explican los fundamentos básicos de la regresión lineal aplicada a un casi sencillo relacionado con la cyberseguridad

### Enunciado del ejercicio
El ejercicio consiste en predecir el costo de un incidente de seguridad en base a un número de equipos que se ven afectados, El DataSet es generado de manera aleatoria.

# 1.- Generacion del DataSet

```python
import numpy as np

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

print("La longitud del DataSet es: ", len(X))
```

# 2.-  Visualizacion del DataSet

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(X, y, "b.")
plt.show()
```

```python
plt.plot(X, y, "b.")
plt.title("Graphics")
plt.xlabel("Equipos afectados (u/1,000)")
plt.ylabel("Costo del incidente (u/10,000)")
plt.show()
```

# 3.- Modificacion del DataSet

```python
import pandas as pd
```

```python
data = {'No_Equipos_Afectados': X.flatten(), 'Costo': y.flatten()}
df = pd.DataFrame(data)
df.head(10)
```

```python
# Escalado del número de equipos afectados.
df['No_Equipos_Afectados'] = df ['No_Equipos_Afectados'] * 1000
df['No_Equipos_Afectados'] = df ['No_Equipos_Afectados'].astype('int')

#Escalado del costo

df['Costo'] = df['Costo'] * 10000
df['Costo'] = df['Costo'].astype('int')
df.head(10)
```

```python
# Representacion grafica del DataSet
plt.plot(df['No_Equipos_Afectados'], df['Costo'], "b.")
plt.title("Graphics")
plt.xlabel("Equipos afectados")
plt.ylabel("Costo del incidente")
plt.show()
```

# 4.- Construccion del modelo 

```python
from sklearn.linear_model import LinearRegression
```

```python
# Construccion del modelo y ajuste de la hipotesis
lin_reg = LinearRegression()
lin_reg.fit(df['No_Equipos_Afectados'].values.reshape(-1, 1), df['Costo'].values)
```

```python
# Parametro tetha 0
lin_reg.intercept_
```

```python
# Parametro tetha 1 coeficiente de determinacion
lin_reg.coef_
```

```python
# Prediccion para el valor minimo y maximo del DataSet 
X_min_max = np.array([[df['No_Equipos_Afectados'].min()], [df['No_Equipos_Afectados'].max()]])
y_train_pred = lin_reg.predict(X_min_max)
```

```python
# Representacion gráfica de la función de hipotesis generada.
plt.plot(X_min_max, y_train_pred, "g-")
plt.plot(df['No_Equipos_Afectados'], df['Costo'], "b.")
plt.xlabel("Equipos Afectados")
plt.ylabel("Costo del incidente")
plt.show()
```

# 5.- Prediccion de nuevos ejemplos 

```python
x_new = np.array([[1255]]) # Numero de equipos Afectados
#Prediccion del costo que tendria el incidente.
Costo = lin_reg.predict(x_new)

print("El costo del incidente seria:", int(Costo[0]))
```

```python
plt.plot(df['Numero_Equipos_Afectados'], df['Costo'], "b.")
plt.plot(X_min_max, y_train_pred, "g-")
plt.plot(x_new, Costo, "W")
plt.xlabel("Equipos Afectados TESJI")
plt.ylabel("Costo del incidente")
plt.show
```
